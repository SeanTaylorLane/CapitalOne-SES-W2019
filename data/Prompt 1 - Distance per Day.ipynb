```python
import pandas as pd
```

```python
import dateutil.parser
```

```python
from geopy import distance
from geopy import Point
```

```python
import math
```

```python
data = pd.read_csv('metro-bike-share-trip-data.csv', dtype={'Trip ID': int,
                                                            'Duration': int,
                                                           'Start Time': str,
                                                           'End Time': str,
                                                           'Starting Station ID': str,
                                                           'Starting Station Latitude': float,
                                                           'Starting Station Longitude': float,
                                                           'Ending Station ID': str,
                                                           'Ending Station Latitude': float,
                                                           'Ending Station Longitude': float,
                                                           'Bike ID': str,
                                                           'Plan Duration': str,
                                                           'Trip Route Category': str,
                                                           'Passholder Type': str,
                                                           'Starting Lat-Long': str,
                                                           'Ending Lat-Long': str})
```

```python
data
```

```python
# Aggregate the total distance traveled on each weekday and the number of trips per weekday
# Index 0 corresponds to Monday while index 6 corresponds to Sunday
distances = [0, 0, 0, 0, 0, 0, 0]
num_trips = [0, 0, 0, 0, 0, 0, 0]
for i, row in data.iterrows():
    start_lat = row['Starting Station Latitude']
    start_long = row['Starting Station Longitude']
    end_lat = row['Ending Station Latitude']
    end_long = row['Ending Station Longitude']
    # Some of the lat/long values in the csv are NaN, so we need to make sure we don't
    # include those in our calculations
    if(math.isnan(start_lat) or math.isnan(start_long) or math.isnan(end_lat) or math.isnan(end_long)):
        continue
    start = Point(start_lat, start_long)
    end = Point(end_lat, end_long)
    # We aren't including round-trip trips
    if(distance.distance(start, end).miles != 0.0):
        date = dateutil.parser.parse(row['Start Time'])
        distances[date.weekday()] += distance.distance(start, end).miles
        num_trips[date.weekday()] += 1
```

```python
average_distances = [0, 0, 0, 0, 0, 0, 0]
for i in range(len(average_distances)):
    if num_trips[i] == 0:
        continue
    average_distances[i] = distances[i]/num_trips[i]
```
